```python
import cv2
import sys

imagePath = sys.argv[1]

image = cv2.imread('img/messi.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.3,
    minNeighbors=3,
    minSize=(30, 30)
)

print("[INFO] Found {0} Faces!".format(len(faces)))

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
#print("[INFO] Image faces_detected.jpg written to filesystem: ", status)   to save on your system
plt.imshow(image,cmap='gray')
```

## LBP cascade classifier

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.3,minNeighbors=3):
    #just making a copy of image passed, so that passed image is not changed
    img_copy = colored_img.copy()
    #convert the test image to gray image as opencv face detector expects gray images
    gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)    
    img_copy=cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB)
    #let's detect multiscale (some images may be closer to camera than others) images
    faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=5);   
    print("[INFO] Found {0} Faces!".format(len(faces)))
    
    #go over list of faces and draw them as rectangles on original colored img
    for (x, y, w, h) in faces:
        cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
        return img_copy
```

```python
#lbp_face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'lbpcascade_frontalface.xml') 
#lbp_face_cascade = cv2.CascadeClassifier('data/lbpcascade_frontalface.xml')  
lbp_face_cascade  = cv2.CascadeClassifier('lbpcascade_frontalface.xml');
```

```python
#load test image
test2 = cv2.imread('img/people_with_phones.png')
#call our function to detect faces
faces_detected_img = detect_faces(lbp_face_cascade, test2)

plt.imshow(faces_detected_img)
```
